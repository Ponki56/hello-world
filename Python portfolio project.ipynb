```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('D:\Data Science Projects\movies.csv')
```

```python
df.head()
```

```python
#checking for missing data
for col in df.columns:
    missing_df = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, missing_df))
         
```

```python
#Data types for columns
df.dtypes
```

```python
df.sort_values(by = ['gross'], inplace=True, ascending = False)
```

```python
df
```

```python
#dropping any duplicates or showing all distinct names
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Budget high correltation

plt.scatter(df['budget'], df['gross'])
plt.xlabel('Movie Budget')
plt.ylabel('Gross Earnings')
plt.title('Budget vs Gross earnings')
plt.show
```

```python
#Budget vs Gross using Regression plot
sns.regplot(x = df['budget'], y = df['gross'], scatter_kws = {'color':'red'},line_kws = {'color':'blue'})
```

```python
#Looking at correlation
df.corr(method = 'pearson')
```

```python
#There seems to be high correlation between budget and gross
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] =df_numerized[col_name].astype('category')
        df_numerized[col_name] =df_numerized[col_name].cat.codes
df_numerized        
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.35]
high_corr
```
